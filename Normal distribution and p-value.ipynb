```python
# what is the probablity of getting 1 Head in 10 coin toss?
# what is the probability of getting 2 Head in 10 coin toss?
```

```python
n = 10
no_of_heads = range(0, n+1) # no of heads ranges from 0 to 10
p = 0.5
q = 1-p
```

```python
from scipy.special import comb
```

```python
probability = [comb(N=n, k=x) * p**x * q ** (n-x) for x in no_of_heads]
probability
```

```python
import plotly.express as px
fig = px.bar(× =no_of_heads, y=probability, text = probability)
fig.update_layout(width = 500,
                  height = 400),
                  xaxis_title = "No. of Heads",
                  yaxis_title = 'probability')
fig.show()
```

#Hypothesis testing

```python
HO = "The coin is Unbiased"
H1 = 'The coin is Biased'
alpha = 0.05
```

```python
# suppose on your test data you've got 9H and IT in 10 total tosses
```

```python
from scipy.stats import binomtest
result = binomtest(k=9, n=10, p=0.5)
result
```

```python
result.pvalue
```

```python
if result.pvalue <= alpha:
  print("We reject HO")
else:
  print("We fail to reject HO")
```
