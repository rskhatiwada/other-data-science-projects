```python
import numpy as np
```

```python
np.random.seed(0)
d1 = np.random.normal(loc=5, scale = 1, size = 1500) # loc = mean, scale = std
d2 = np.random.normal(loc=5, scale = 10, size = 1500)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.kdeplot(d1, fill = True, label = 'd1')
sns.kdeplot(d2, fill = True, label = 'd2')
plt.axvline(5, color = 'red', label = 'mean')
plt.legend()
plt.show()
```

```python
#normal distribution vs standard distribution.ipynb
```

Normal:
* mean = $\mu$
* Variance = $\sigma^2$

Standard Normal Distribution
* mean = 0
* Variance = 1

Z-Score is used to convert Normal Distribution to Standard Normal Distribution

$Z-Score = \frac{x_i - \mu}{\sigma}$
