# T Test

A t-test is a type of inferential statistic which is used to determine if there is a significant difference between the means of two groups which may be related in certain features

T-test has 2 types : 1. one sampled t-test 2. two-sampled t-test.

# One Sample T-Test with Python
The test tells us whether means of the sample and the population are same.

```python
import numpy as np
np.random.seed(0)
population_ages = np.random.randint(1, 100, size=150)
print(population_ages)
mean_age = np.mean(population_ages)
print("Mean age", mean_age)
```

# Whether there is a significant differencre between sample and population mean





*   H0: There is no difference in population and sample mean
*   H1: There is significant difference between sample and population mean


One Sample T-Test

```python
from scipy import stats
```

```python
result = stats.ttest_1samp(population_ages, popmean = mean_age)
result
```

```python
alpha = 1.0
if result.pvalue < alpha:
  print("We reject H0")
else:
  print("We fail to reject H0")
```

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head(2)
```

# Is there any relationship between sex and smoking?



*   H0: There is no relationship
*   H1: There is some relationship
alpha = 0.05



# Cross Tabulation

```python
import pandas as pd
table = pd.crosstab(df.sex, df.smoker)
table
```

```python
table.values
```

```python
from scipy.stats import chi2_contingency
```

```python
result = chi2_contingency(table.values)
result
```

```python
if result.pvalue < alpha:
  print("We Reject h0")
else:
  print("We fail to reject H0")
```
